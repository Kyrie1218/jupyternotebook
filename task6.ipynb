*(i)在教材例18.7中，我们估计了六月期国库券持有期收益率的一个误差修正模型，其中三月期国库券持有期收益率的一阶滞后为解释变量。我们假定在方程$hy6_{t}=\alpha+\beta hy3_{t-1}+u_{t}$中的协整参数为1。现在，添加$\Delta hy3_{t-1}$的先导变化$\Delta hy3_{t}$、同期变化$\Delta hy3_{t-1}$和滞后变化$\Delta hy3_{t-2}$。即估计方程
$$hy6_{t}=\alpha+\beta hy3_{t-1}+\phi_{0}\Delta hy3_{t}+\phi_{1}\Delta hy3_{t-1}+\rho_{1}\Delta hy3_{t-2}+e_{t}$$
并用方程形式报告结果。相对于双侧备择假设，检验$H_{0}:\beta=1$。假定方程中已经有了足够多的先导和滞后，使得{$hy3_{t-1}$}在这个方程中是严格外生的，我们不必担心序列相关。*

```r
#载入用于时间序列分析的R包
library(dynlm)
library(zoo)
#回归
res<-dynlm(hy6~L(hy3)+d(hy3)+L(d(hy3))+L(d(hy3),2),data=intqrt.ts)
```

```r
#载入用于展示回归结果的R包
library(stargazer)
```

```r
#展示回归结果
stargazer(res,type="text")
```

```r
#检验$H_{0}:\beta=1$
n=121#样本量
b<-summary(res)$coef[2]#beta的估计系数
b0<-1#原假设的beta系数
se<-summary(res)$coef[2,2]#beta估计系数的标准误
t<-(b-b0)/se#检验统计量
pvalue<-2*pt(-abs(t),df=n-1)#p值计算公式
pvalue#p值
```

```r
#分别在alpha=0.01,0.05，0.10下进行假设检验
(pvalue<=0.01)
(pvalue<=0.05)
(pvalue<=0.10)
```

*(ii)在教材(18.39)中的误差修正模型中，添加$\Delta hy3_{t-2}$和$(hy6_{t-2}-hy3_{t-3})$。这两项是联合显著的吗？你认为怎样才是适当的误差修正模型？*

```r
res1<-dynlm(d(hy6)~L(d(hy3))+L(hy6hy3_1),data=intqrt.ts)#教材原模型回归
res2<-dynlm(d(hy6)~L(d(hy3))+L(hy6hy3_1)+L(d(hy3),2)+L(hy6hy3_1,2),data=intqrt.ts)#新模型回归
stargazer(res1,res2,type="text")#展示两个回归的结果
```

```r
#进行新加项的联合检验
library(carData)
library(car)#载入联合检验所需的R包
linearHypothesis(res2,c("L(d(hy3), 2)=0","L(hy6hy3_1, 2)=0"))#对新加项进行联合F检验
```

**解释：**
- 联合检验在显著性水平0.10下不显著；
- 教材中原模型较为适当（新模型新加的变量联合不显著，且未显著提高调整R方）

```r
#加入gwage314的一阶滞后项,检验Granger因果关系
res1<-dynlm(gemp314~L(gemp314)+L(gwage314),data=minwage.ts)
stargazer(res1,type="text")
```

- L（gwage314）系数的统计显著性：在0.01显著性水平下，拒绝原假设，即拒绝“gwage314不是gemp314的格兰杰原因”的原假设

*(iii)向模型中加入就业增长的二阶滞后项，并做格兰杰因果检验。你的结论是否改变？*

```r
#再加入gwage314的二阶滞后项,检验Granger因果关系
res2<-dynlm(gemp314~L(gemp314)+L(gwage314)+L(gwage314,2),data=minwage.ts)
stargazer(res2,type="text")
```
